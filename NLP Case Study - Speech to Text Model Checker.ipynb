```python
import numpy as np
import sounddevice as sd
import soundfile as f
import librosa
import IPython.display as ipd
import warnings
import tensorflow as tf
warnings.filterwarnings("ignore")
from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam
from scipy.io.wavfile import write
from scipy.io import wavfile
```

```python
print(tf.__version__)
```

```python
outcome_class=['down',
 'eight',
 'five',
 'four',
 'left',
 'nine',
 'no',
 'one',
 'right',
 'seven',
 'six',
 'three',
 'two',
 'up',
 'yes']
```

```python
model=load_model('best_model_using_LSTMs_epochs_200_samples_2000.hdf5')
```

```python
model.summary()
```

```python
def predict(audio_s):
    probability=model.predict(audio_s.reshape(1,8000,1))
    index=np.argmax(probability[0])
    return outcome_class[index],index
```

```python
def predict_file(filename):
    samplerate, data = wavfile.read(filename)
    data=librosa.resample(data,16000,8000)
    return predict(data)
```

```python
filename="my_voice/left.wav"
```

```python
ipd.Audio(filename)
```

```python
predict_file(filename) 
```

```python
sample_rate=16000
duration=1
fname="my_voice/right.wav"
```

```python
print("Started Recording")
voice=sd.rec(int(sample_rate*duration),samplerate=sample_rate,channels=1,blocking=True)
print("Stopped Recording")
sd.wait()
```

```python
samples=librosa.resample(voice.ravel(),sample_rate,8000)
predicted_class,index=predict(samples)
print(predicted_class)
```

```python
write(fname,sample_rate,voice)
```
